## Library

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

## Select tickers and the dates

```python
# Change this part only then hit shift+enter
# ---------------------------------------------------
ticker = 'AAPL'
benchmark = 'SPY'
start = '01/01/2010'
end = '01/01/2016'
#----------------------------------------------------
```

## Daily Beta

```python
df = data.DataReader(ticker, 'yahoo', start, end)
bm = data.DataReader(benchmark, 'yahoo', start, end)
df['comp_ret'] = df['Adj Close'].pct_change()
df['bm_ret'] = bm['Adj Close'].pct_change()
df.dropna(inplace = True)
lm.fit(np.array(df['bm_ret']).reshape(-1, 1), np.array(df['comp_ret']).reshape(-1, 1))
x1 = print('Coefficent :', lm.coef_)
x2 = plt.scatter(df['bm_ret'], df['comp_ret'])
plt.xlabel(benchmark)
plt.ylabel(ticker)
print(x1) 
```

## Weekly Beta

```python
df_w = data.get_data_yahoo(ticker, start, end, interval ='w')
bm_w = data.get_data_yahoo(benchmark, start, end, interval ='w')
df_w['comp_ret'] = df_w['Adj Close'].pct_change()
df_w['bm_ret'] = bm_w['Adj Close'].pct_change()
df_w.dropna(inplace = True)
lm.fit(np.array(df_w['bm_ret']).reshape(-1, 1), np.array(df_w['comp_ret']).reshape(-1, 1))
x1 = print('Coefficent :', lm.coef_)
x2 = plt.scatter(df_w['bm_ret'], df_w['comp_ret'])
plt.xlabel(benchmark)
plt.ylabel(ticker)
x1 
```
